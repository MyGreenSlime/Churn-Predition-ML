```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
forcast = pd.read_csv('forcast.csv')
```

```python
forcast.head()
```

```python
forcast['SHOP_DATE'] = pd.to_datetime(forcast['SHOP_DATE'])
```

```python
forcast.set_index('SHOP_DATE', inplace=True)
```

```python
forcast.plot(figsize=(30,10), linewidth=5, fontsize=20)
```
