```python
def mostcutsen(CUS): #la 0 mm 1 um 2 xx 3
    MAXX = max([CUS['CUS_SEN_LA'],CUS['CUS_SEN_MM'],CUS['CUS_SEN_UM'],CUS['CUS_SEN_XX']])
    if MAXX == CUS['CUS_SEN_LA']:
        return 0
    elif MAXX == CUS['CUS_SEN_MM']:
        return 1
    elif MAXX == CUS['CUS_SEN_UM']:
        return 2
    elif MAXX == CUS['CUS_SEN_XX']:
        return 3
```

```python
def mostbassize(CUS): #L 0 M 1 S 2
    MAXX = max([CUS['BAS_SIZE_L'],CUS['BAS_SIZE_M'],CUS['BAS_SIZE_S']])
    if MAXX == CUS['BAS_SIZE_L']:
        return 0
    elif MAXX == CUS['BAS_SIZE_M']:
        return 1
    elif MAXX == CUS['BAS_SIZE_S']:
        return 2
```

```python
def mostbasdo(CUS): #fresh 0 grocery 1 mix 2 nonfood 3
    MAXX = max([CUS['BAS_DO_FRESH'],CUS['BAS_DO_GROCERY'],CUS['BAS_DO_MIX'],CUS['BAS_DO_NONFOOD']])
    if MAXX == CUS['BAS_DO_FRESH']:
        return 0
    elif MAXX == CUS['BAS_DO_GROCERY']:
        return 1
    elif MAXX == CUS['BAS_DO_MIX']:
        return 2
    elif MAXX == CUS['BAS_DO_NONFOOD']:
        return 3
```

```python
def moststorefor(CUS): #xx 0 ls 1 ms 2 xls 3
    MAXX = max([CUS['STORE_FOR_XX'],CUS['STORE_FOR_LS'],CUS['STORE_FOR_MS'],CUS['STORE_FOR_XLS']])
    if MAXX == CUS['STORE_FOR_XX']:
        return 0
    elif MAXX == CUS['STORE_FOR_LS']:
        return 1
    elif MAXX == CUS['STORE_FOR_MS']:
        return 2
    elif MAXX == CUS['STORE_FOR_XLS']:
        return 3
```

```python
def mostbastype(CUS): #full 0 small 1 top 2 xx 3
    MAXX = max([CUS['BAS_TYPE_FULL'],CUS['BAS_TYPE_SMALL'],CUS['BAS_TYPE_TOP'],CUS['BAS_TYPE_XX']])
    if MAXX == CUS['BAS_TYPE_FULL']:
        return 0
    elif MAXX == CUS['BAS_TYPE_SMALL']:
        return 1
    elif MAXX == CUS['BAS_TYPE_TOP']:
        return 2
    elif MAXX == CUS['BAS_TYPE_XX']:
        return 3
```
